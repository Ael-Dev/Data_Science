# **PARTE 2**
Utilizando datos solo del Departamento/Regi√≥n del Cusco

```python
# separamos los datos de la Regi√≥n del Cusco utilizando:
# Qosqo = ............
Qosqo = Datos_df[Datos_df['Departamento'] == 'Cusco'].copy()
Qosqo=Qosqo.reset_index(drop=True) # reestablecer el indice(opcional)
Qosqo
```

```python
# Qosqo es una estructura de?    ponga una X en la(s) que considere corresta(s)
#    Dataframe (X)  type(Qosqo)
#    Serie     ( )
```

```python
# cambie los rotulos de Qosqo las columnas (Departamento x a√±o  y  Cusco x Poblaci√≥n)
# .........
Qosqo1 = Qosqo.rename(columns={'Departamento':'A√ëO'}) #cambiar columna
Qosqo_df = pd.DataFrame(Qosqo1)
Qosqo_df.loc[Qosqo_df.A√ëO == "Cusco","A√ëO"] = 'Poblacion' #cambiar fila
Qosqo_df
```
