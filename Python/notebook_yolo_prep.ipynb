En este ejemplo, se puede ver que el recall es más importante que la precision, ya que lo que se quiere evitar es dejar de diagnosticar a las personas que tienen la enfermedad. Se debe usar la precision cuando se quiere minimizar los falsos positivos, es decir, cuando se quiere tener una alta confianza en las predicciones positivas. Por ejemplo, en la detección de fraudes, se quiere evitar que las transacciones legítimas se marquen como fraudulentas1.
