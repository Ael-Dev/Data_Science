## DATA AUGMENTATION - MIRROR and ROTATION

```python
# data augmentation IMAGEN ESPEJO

# Importar las librerías necesarias
import cv2
import glob
import os

# Obtener la lista de las imágenes de la carpeta data
images = glob.glob("./dataset/positivo/*.jpg")

# Definir la ruta de la carpeta nueva
new_folder = "data_augmented"

# Crear la carpeta nueva si no existe
if not os.path.exists(new_folder):
    os.mkdir(new_folder)

# Iterar sobre la lista de imágenes
for i, image in enumerate(images):
    # Cargar la imagen
    img = cv2.imread(image)
    

    # Crear una imagen espejo horizontal de la imagen original
    img_mirror = cv2.flip(img, 1)
    # crear una imagen rotada de la imagen espejo
    img_mirror_rotate = cv2.rotate(img_mirror, cv2.ROTATE_180)

    # crear una imagen rotado 180 grados
    img_rotate_180 = cv2.rotate(img, cv2.ROTATE_180)
    # Crear una imagen espejo horizontal de la imagen rotada
    img_rotate_mirror = cv2.flip(img_rotate_180, 1)


    # Obtener el nombre de la imagen
    name = os.path.splitext(os.path.basename(image))[0]

    # Guardar la imagen espejo en la carpeta nueva
    cv2.imwrite(f"./{new_folder}/{name}_m.jpg", img_mirror)
    # Guardar la imagen espejo en la carpeta nueva
    cv2.imwrite(f"./{new_folder}/{name}_ro.jpg", img_rotate_180)
    # Guardar la imagen espejo de la rotacion en la carpeta nueva
    #cv2.imwrite(f"./{new_folder}/{name}_ro_m.jpg", img_rotate_mirror)
    # Guardar la imagen rotada-del espejo en la carpeta nueva
    cv2.imwrite(f"./{new_folder}/{name}_m_ro.jpg", img_mirror_rotate)

    print(i+1)
```

## USANDO TENSORFLOW

```python
# Importar las librerías necesarias
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Definir la carpeta que contiene las imágenes
folder = "./dataset/positivo"

# Definir los parámetros de la data augmentation
datagen = ImageDataGenerator(
    rotation_range = 20, # rotar las imágenes hasta 20 grados
    width_shift_range = 0.2, # desplazar las imágenes horizontalmente hasta el 20% de su ancho
    height_shift_range = 0.2, # desplazar las imágenes verticalmente hasta el 20% de su alto
    shear_range = 0.2, # aplicar un corte de hasta 20 grados
    zoom_range = 0.2, # aplicar un zoom de hasta el 20%
    horizontal_flip = True, # voltear las imágenes horizontalmente
    fill_mode = "nearest" # rellenar los píxeles vacíos con el valor más cercano
)

# Crear una carpeta para guardar las imágenes generadas
output_folder = "augmented_images"
if not os.path.exists(output_folder):
    os.mkdir(output_folder)

# Leer las imágenes de la carpeta original y aplicar la data augmentation
for filename in os.listdir(folder):
    # Cargar la imagen y convertirla a un array de numpy
    image = cv2.imread(os.path.join(folder, filename))
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = np.expand_dims(image, axis = 0)

    # Generar 10 imágenes nuevas a partir de la original
    i = 0
    for batch in datagen.flow(image, batch_size = 1, save_to_dir = output_folder, save_prefix = filename.split(".")[0], save_format = "jpg"):
        i += 1
        if i == 10:
            break

print("La data augmentation se ha aplicado con éxito sobre la carpeta", folder)
```

# RESIZE IMAGES

```python
from PIL import Image
import glob
import os

# funcion para redimensionar imagenes
def resize_images_640(images_path_glob, new_folder):
    # Crear la carpeta nueva si no existe
    if not os.path.exists(new_folder):
        os.mkdir(new_folder)
    # Iterar sobre la lista de imágenes
    for i, image in enumerate(images_path_glob):
        # obtener el nombre de la imagen
        name = os.path.basename(image)[:-4]
        # Cargar la imagen
        img = Image.open(image)
        # Redimensionar la imagen a 640x640
        img = img.resize((640, 640))
        # Guardar los recortes en la carpeta new_folder
        img.save(f"./{new_folder}/{name}.jpg")
        # # mostrar el nro de imagen a recortar   
        print(f'croping image {i+1}/{len(images_path_glob)}')


# Obtener la lista de las imágenes de la carpeta dataset
images_positivo = glob.glob("./data_augmented/positivo/*.jpg")
images_negativo = glob.glob("./data_augmented/negativo/*.jpg")

# Definir la ruta de la carpeta nueva
new_folder1 = "positivo_640x640"
new_folder2 = "negativo_640x640"
# resize
resize_images_640(images_positivo,new_folder1)
resize_images_640(images_negativo,new_folder2)
```
