```python
# @title Set the date/time column as the index

# Copiar el dataframe original
df_comp = df.copy()

# Convertir la columna 'Date/Time' a tipo datetime
df_comp['Date/Time'] = pd.to_datetime(df_comp['Date/Time'], format='%m/%d/%Y %H:%M', dayfirst=True)

# Poner como índice la columna 'date/time' 
df_comp.set_index('Date/Time', inplace=True)

# Aplicar una frecuencia de días laborables al índice
df_comp = df_comp.asfreq('H')

# Rellenar los valores faltantes con el método 'ffill'
df_comp = df_comp.fillna(method='ffill')

# Mostrar las primeras filas del dataframe modificado
df_comp.head()
```

```python
df_comp.hist();
```

```python
# @title Obtain categorical and numerical columns
# ----------------------------------------------------------------
categorical_features = df_comp.select_dtypes(include='object').columns
numerical_features = df_comp.select_dtypes(exclude='object').columns

print('categorical features:', categorical_features)
print('numerical features:', numerical_features)
```

```python
# @title Plot features vs date_time
df_comp.plot(subplots=True, figsize=(14, 10), style='-')
# Mostrar el gráfico
plt.show()
```

```python
# @title Scatter plot
# ----------------------------------------------------------------
g = sns.pairplot(df_comp, diag_kind="kde")
g.map_lower(sns.kdeplot, levels=4, color=".2")
```

```python
# @title Correlation matrix
# ----------------------------------------------------------------
corr = df_comp.iloc[:,:-1].corr()
sns.heatmap(corr, annot=True, fmt=".2f", linewidths=0.5)
```
