# <font color=blue>Visualizacion con Matplotlib</font>

La libreria Matplotlib es una de las librerias de visualizacion mas usadas en Python en tareas de Machine Learning.
Uno de los modulos centrales de Matplotlib es <b>matplotlib.pyplot</b>. Este provee una colección de funciones al estilo de comandos. Cada función de pyplot realiza algún cambio en una figura: por ejemplo, crea una figura, crea un área de trazado en una figura, traza algunas líneas en un área de trazado, decora el trazado con etiquetas, etc. 


```python
# importar matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python
# Genere el plot de las 3 columnas de cursos vs columna 'Mes' 
# incluya en el grafico toda la información que corresponda

plt.figure(figsize=[20,15])   #tamaño figura 


# incluir otros parametros
# incluir otros parametros
plt.plot(data["Mes"],data["data science"], label="data science", color="red", marker=".", linestyle="-.")
plt.plot(data["Mes"],data["machine learning"],label="machine learning", color="blue", marker="+", linestyle="--")
plt.plot(data["Mes"],data["deep learning"],label="deep learning", color="green", marker="<", linestyle=":")

plt.legend()                  # Coloca leyenda
plt.title("Cursos por mes")   # Coloca titulo
plt.xlabel("Mes")             # Coloca etiqueta en eje de absisas
plt.ylabel("Cursos")          # Coloca etiqueta en eje de ordenas
plt.yticks(np.arange(0, 101, 5))                                # Habilita el grillado en el plot
plt.xticks(np.arange(0, 193, 12))                               # Habilita el grillado en el plot
plt.grid()
plt.minorticks_on()
plt.show()
```

# **<font color=blue>Homogenización de Datos</font>**

Llamado tambien re-escalamiento de datos, está en la etapa de preprocesamiento de datos.

La operacion de re-escalamiento transforma los atributos de entrada a un mismo intervalo.

Importando las funciones de la libreria scikit-learn que hacen las operaciones de re-escalamiento: 


*   MinMaxScaler
*   Normalizer
*   StandardScaler
*   RobustScaler

Homogenizar (poner en la misma escala) el último data frame, para cada caso:

```python
from sklearn import preprocessing
```

```python
datos = pd.read_csv("pima-indians-diabetes.data.csv",names=col_names)
datos
```

Variables:

preg = embarazo

plas = glucosa

pres = presion arterial

skin = Grosor de la piel

test = Insulina

mass= indice de masa corporal

pedi= pedigree de diabetes

age = edad

class = Variable de clase (0 o 1) 268 de 768 son 1, los demás son 0

## ***Escala en función del mínimo y máximo***


```python
# utilizando MinMaxxScaler

datos_min_max = preprocessing.MinMaxScaler().fit_transform(data)
datos_min_max
```

## ***Normaliza en función de la Norma del Vector***# 

```python

# utilizando la norma del vector del data frame, llamado Normalizer

datos_normalizer = preprocessing.Normalizer().transform(data.T)
datos_normalizer = datos_normalizer.T
# normalizado = x / raiz_cuadrada(x_1^2 + x_2^2 + x_3^2 + ... )

print(data)     # <-- datos originales
datos_normalizer # <-- datos normalizados
```

## ***Estandariza (desv_std=1, media=0)***

```python
# utilizando la desviación estandar con valor 1 y la media con valor 0

datos_standar_scaler = preprocessing.StandardScaler().fit_transform(datos)
# estandarizado = (X - media) / std

datos_standar_scaler
```

##**ROBUST-SCALER**

```python
# utilizando el escalador llamado robusto. Utiliza el 50% de los datos centrales

datos_robust_scaler = preprocessing.RobustScaler().fit_transform(datos)
# estandarizado = (X - rango_intercuartílico) / std

datos_robust_scaler
```

# **<font color=blue>Visualiación de Datos**</font>

Con los datos obtenidos en el procediemiento anterior y previa conversión de vectores de numpy a DataFrames, gráfique las cuatro homogenizaciones obtenidas.


```python
# convierte vectores de numpy a DataFrames para graficarlos
Encabezado = data.columns.tolist()
datos_min_max        = pd.DataFrame(datos_min_max,        columns= Encabezado)
datos_normalizer     = pd.DataFrame(datos_normalizer,     columns= Encabezado)
datos_standar_scaler = pd.DataFrame(datos_standar_scaler, columns= Encabezado)
datos_robust_scaler  = pd.DataFrame(datos_robust_scaler,  columns= Encabezado)

# crea una figura con 5 subfiguras para comparar los métodos
fig = plt.figure(figsize=(15, 3))
ax1 = fig.add_subplot(1, 5, 1)
ax2 = fig.add_subplot(1, 5, 2)
ax3 = fig.add_subplot(1, 5, 3)
ax4 = fig.add_subplot(1, 5, 4)
ax5 = fig.add_subplot(1, 5, 5)

# construye y personaliza series de datos

# con datos originales
ax1.set_title("Original-age")
ax1.plot(data["age"], linewidth=0, marker="*", color="green", markersize=4)

# con datos estandarizados utilizando MinMaxScaler
ax2.set_title("Min Max")
ax2.plot(datos_min_max["age"], linewidth=0, marker="*", color="green", markersize=4)

# con datos estandarizados utilizando Normalizer
ax3.set_title("Normalizer")
ax3.plot(datos_normalizer["age"], linewidth=0, marker="*", color="green", markersize=4)

# con datos estandarizados utilizando StandarScaler
ax4.set_title("Standard Scaler")
ax4.plot(datos_standar_scaler["age"], linewidth=0, marker="*", color="green", markersize=4)

# con datos estandarizados utilizando RobustScaler
ax5.set_title("Robust Scaler")
ax5.plot(datos_robust_scaler["age"], linewidth=0, marker="*", color="green", markersize=4)

plt.show()
```

```python
# convierte vectores de numpy a DataFrames para graficarlos
Encabezado = data.columns.tolist()
datos_min_max        = pd.DataFrame(datos_min_max,        columns= Encabezado)
datos_normalizer     = pd.DataFrame(datos_normalizer,     columns= Encabezado)
datos_standar_scaler = pd.DataFrame(datos_standar_scaler, columns= Encabezado)
datos_robust_scaler  = pd.DataFrame(datos_robust_scaler,  columns= Encabezado)

# crea una figura con 5 subfiguras para comparar los métodos
fig = plt.figure(figsize=(15, 3))
ax1 = fig.add_subplot(1, 5, 1)
ax2 = fig.add_subplot(1, 5, 2)
ax3 = fig.add_subplot(1, 5, 3)
ax4 = fig.add_subplot(1, 5, 4)
ax5 = fig.add_subplot(1, 5, 5)

# construye y personaliza series de datos
variable = 'mass'
# con datos originales
ax1.set_title("Original")
ax1.plot(data[variable], linewidth=0, marker="*", color="purple", markersize=4)

# con datos estandarizados utilizando MinMaxScaler
ax2.set_title("Min Max")
ax2.plot(datos_min_max[variable], linewidth=0, marker="*", color="purple", markersize=4)

# con datos estandarizados utilizando Normalizer
ax3.set_title("Normalizer")
ax3.plot(datos_normalizer[variable], linewidth=0, marker="*", color="purple", markersize=4)

# con datos estandarizados utilizando StandarScaler
ax4.set_title("Standard Scaler")
ax4.plot(datos_standar_scaler[variable], linewidth=0, marker="*", color="purple", markersize=4)

# con datos estandarizados utilizando RobustScaler
ax5.set_title("Robust Scaler")
ax5.plot(datos_robust_scaler[variable], linewidth=0, marker="*", color="purple", markersize=4)

plt.show()
```
